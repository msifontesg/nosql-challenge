### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01

#Filter by business name
penang_query = {"BusinessName":"Penang Flavours"}
penang_results = establishments.find_one(penang_query)
#Get lat/lon
latitude = penang_results.get('geocode').get('latitude')
longitude = penang_results.get('geocode').get('longitude')
#Apply degrees of search
min_lat = latitude - degree_search
max_lat = latitude + degree_search
min_lon = longitude - degree_search
max_lon = longitude + degree_search


query = {'$and':[{'geocode.latitude':{'$gte': min_lat}},{'geocode.latitude':{'$lte':max_lat}}
                 ,{'geocode.longitude':{'$gte':min_lon}},{'geocode.longitude':{'$lte':max_lon}}]}
sort =  [('scores.Hygiene', 1)]

# Print the results
result3 = establishments.find(query).sort(sort).limit(5)
for result in result3:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
result3 = establishments.find(query).sort(sort).limit(5)
top5_df = pd.DataFrame(result3).set_index('_id',drop=True)
top5_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_4 = {'$match':{'scores.Hygiene': 0} }
# 2. Groups the matches by Local Authority
group_4 = {'$group': {'_id': "$LocalAuthorityName", 'count': {'$sum' : 1}}}
# 3. Sorts the matches from highest to lowest
sort_4 = {'$sort': { 'count': -1 }}

local_auth = [match_4, group_4, sort_4]
result4 = list(establishments.aggregate(local_auth))

# Print the number of documents in the result
print(f"Number of documents in the result: {len(result4)}")

# Print the first 10 results
pprint(result4[0:10])
```

```python
# Convert the result to a Pandas DataFrame
result4_df = pd.DataFrame(result4)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {result4_df.shape[0]}")

# Display the first 10 rows of the DataFrame
result4_df.head(10)
```
